```python
import pandas as db

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/arletteGuerrero12/Challenge_TELECOMX_LATAM/refs/heads/main/TelecomX_Data.json"
df = pd.read_json(url)
# Mostrar las primeras 5 filas del dataset
df.head()
```

```python
df_flat = pd.json_normalize(df.to_dict(orient='records'))
df_flat.columns = df_flat.columns.str.lower().str.replace('.', '_')

print("Columnas del dataframe:")
print(df_flat.columns.tolist())
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder

# Define numerical columns
num_cols = ['account_charges_monthly', 'account_charges_total', 'customer_tenure']

# Convertir columnas numéricas con map y manejo de errores
df_flat[num_cols] = df_flat[num_cols].apply(lambda x: pd.to_numeric(x, errors='coerce'))

# Imputar valores nulos en 'account_charges_total' con SimpleImputer (mediana)
imp_median = SimpleImputer(strategy='median')
df_flat[['account_charges_total']] = imp_median.fit_transform(df_flat[['account_charges_total']])

# Reemplazar servicios inexistentes con "No" usando map
replacements = {"No internet service": "No", "No phone service": "No"}
df_flat = df_flat.apply(lambda col: col.map(replacements).fillna(col) if col.dtype == 'object' else col)

# Crear nueva feature: ratio de cargos vs. tiempo del cliente
df_flat['charges_ratio'] = df_flat.apply(lambda row: row['account_charges_total'] / (row['customer_tenure'] + 1), axis=1)

# Codificar variables categóricas con OrdinalEncoder (alternativa a LabelEncoder)
cat_cols = df_flat.select_dtypes(include='object').columns.tolist()
cat_cols.remove('customerid') # Exclude customerid from encoding
encoder = OrdinalEncoder()
df_encoded = df_flat.copy()
df_encoded[cat_cols] = encoder.fit_transform(df_encoded[cat_cols])

# Vista previa
df_encoded.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ajustes globales
def configurar_estilo():
    sns.set_theme(style="whitegrid")
    plt.rcParams['figure.figsize'] = (10, 6)

# 1. Gráfico de distribución de churn
def plot_churn_distribution(data):
    sns.countplot(data=data, x='churn', palette='pastel')
    plt.title("Distribución de clientes que cancelaron el servicio (Churn)")
    plt.show()

# 2. Gráfico de churn por tipo de contrato
def plot_churn_by_contract(data):
    sns.countplot(data=data, x='account_contract', hue='churn', palette='Set2')
    plt.title("Churn por tipo de contrato")
    plt.xticks(rotation=0)
    plt.show()

# 3. Boxplot de cargos mensuales por churn
def plot_monthly_charges_by_churn(data):
    sns.boxplot(data=data, x='churn', y='account_charges_monthly', palette='Set3')
    plt.title("Distribución de cargos mensuales según Churn")
    plt.show()

# 4. Tenure promedio por tipo de contrato
def plot_avg_tenure_by_contract(data):
    sns.barplot(data=data, x='account_contract', y='customer_tenure', estimator='mean',
                errorbar='sd', palette='muted')
    plt.title("Tenure promedio según contrato")
    plt.ylabel("Tenure (meses)")
    plt.show()

# 5. Matriz de correlación
def plot_correlation_matrix(data, cols):
    corr = data[cols].corr()
    sns.heatmap(corr, annot=True, cmap='coolwarm')
    plt.title("Matriz de correlación entre variables numéricas")
    plt.show()

# Ejecutar todos los gráficos
configurar_estilo()
plot_churn_distribution(df_flat)
plot_churn_by_contract(df_flat)
plot_monthly_charges_by_churn(df_flat)
plot_avg_tenure_by_contract(df_flat)
plot_correlation_matrix(df_flat, num_cols)
```

```python
from sklearn.feature_selection import mutual_info_classif
import pandas as pd
import matplotlib.pyplot as plt

# Separar variables predictoras y variable objetivo
X_all = df_encoded.drop(columns=['churn', 'customerid'])
y_target = df_encoded['churn']

# Calcular la información mutua para cada variable
mi_scores = mutual_info_classif(X_all, y_target, discrete_features='auto')
mi_df = pd.DataFrame({'Feature': X_all.columns, 'MI Score': mi_scores})
mi_df.sort_values(by='MI Score', ascending=False, inplace=True)

# Visualizar las 15 principales características
def plot_top_features(mi_dataframe, top_n=15):
    top_features = mi_dataframe.head(top_n)
    plt.figure(figsize=(10, 6))
    plt.barh(top_features['Feature'], top_features['MI Score'], color='skyblue')
    plt.title("Importancia de variables para predecir Churn (Mutual Information)")
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()

# Mostrar el gráfico
plot_top_features(mi_df)

# Seleccionar las variables más relevantes (Top 15)
selected_features = mi_df.head(15)['Feature'].tolist()
X_selected = df_encoded[selected_features]
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report
from imblearn.over_sampling import SMOTE
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# División de datos
X_train, X_test, y_train, y_test = train_test_split(X_selected, y_target, test_size=0.2, random_state=42)

# Escalar variables numéricas seleccionadas
def escalar_variables(X_train, X_test, columnas):
    scaler = StandardScaler()
    X_train[columnas] = scaler.fit_transform(X_train[columnas])
    X_test[columnas] = scaler.transform(X_test[columnas])
    return X_train, X_test

num_vars = [var for var in selected_features if var in num_cols or var == 'charges_ratio']
X_train, X_test = escalar_variables(X_train, X_test, num_vars)

# Balanceo con SMOTE
def balancear_clases(X, y):
    smote = SMOTE(random_state=42)
    return smote.fit_resample(X, y)

X_train_bal, y_train_bal = balancear_clases(X_train, y_train)

# Entrenamiento de modelo
modelo_rf = RandomForestClassifier(n_estimators=100, random_state=42)
modelo_rf.fit(X_train_bal, y_train_bal)

# Predicción y evaluación
y_pred = modelo_rf.predict(X_test)
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=np.unique(y_test),
            yticklabels=np.unique(y_test))
plt.title("Matriz de Confusión")
plt.xlabel("Predicción")
plt.ylabel("Valor real")
plt.tight_layout()
plt.show()

print("\n📊 Reporte de Clasificación:")
print(classification_report(y_test, y_pred))

# Identificar falsos positivos (clientes con riesgo elevado)
X_test_eval = X_test.copy()
X_test_eval['predicted_churn'] = y_pred
X_test_eval['true_churn'] = y_test.values
clientes_riesgo = X_test_eval[(X_test_eval['predicted_churn'] == 1) & (X_test_eval['true_churn'] == 0)]
print(f"Clientes con alto riesgo de cancelar: {clientes_riesgo.shape[0]}")
clientes_riesgo.head()

# Validación cruzada
scores = cross_val_score(modelo_rf, X_selected, y_target, cv=5, scoring='f1_macro')
print(f"🔄 F1-score promedio en validación cruzada: {scores.mean():.3f}")
```

# **INFORME FINAL**


Tras el análisis de datos y la implementación de modelos de machine learning, se han identificado patrones clave que permiten comprender y anticipar la cancelación de clientes:

📌 Hallazgos principales
- El churn está vinculado principalmente a:
- Poca permanencia del cliente (tenure bajo)
- Uso de fibra óptica
- Facturación electrónica y métodos de pago digitales
- Contratos mensuales, sin compromiso a largo plazo
- Cargos mensuales elevados con escasa lealtad
- Factores que disminuyen el riesgo:
- Contratos de 1-2 años
- Servicios adicionales como soporte técnico y seguridad en línea
- Relación más prolongada con la empresa

🎯 Perfil de cliente en riesgo
- Nuevos clientes con poco tiempo en la compañía
- Contratos mensuales
- Uso intensivo del servicio sin beneficios adicionales
- Altos cargos mensuales sin soporte técnico

💡 Estrategias de retención sugeridas
- Fidelización temprana de clientes con bajo tenure
- Descuentos y beneficios para quienes presentan cargos altos y poco tiempo de permanencia
- Migración a contratos de largo plazo con incentivos (descuentos, soporte gratuito)
- Seguimiento proactivo a clientes con fibra óptica y alta facturación
- Campañas dirigidas a clientes con facturación digital pero sin servicios complementarios

🛠️ Recomendaciones técnicas
- Aplicar codificación de variables categóricas
- Estandarizar variables numéricas
- Balancear clases para mejorar el modelo
- Eliminar atributos irrelevantes para optimizar el análisis
